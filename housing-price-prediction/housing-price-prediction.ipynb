```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data, target = fetch_california_housing(as_frame=True, return_X_y=True)
data
```

```python
target
```

```python
df = data.copy()
df
```

```python
df['med_house_value'] = target
df
```

# Data Cleaning and Preprocessing

```python
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={
    "MedInc": "med_income", 
    "HouseAge": "house_age", 
    "AveRooms": "avg_rooms", 
    "AveBedrms": "avg_bedrooms", 
    "Population": "population",
    "AveOccup": "avg_occupancy",
    "Latitude": "latitude",
    "Longitude": "longitude"})
df.head()
```

```python
df.isna().sum()
```

```python
# Drop the unwanted columns
columns_drop = ["longitude", "latitude"]
subset = df.drop(columns=columns_drop)

# Quantize the target and keep the midpoint for each interval
subset["med_house_value"] = pd.qcut(subset["med_house_value"], 6, retbins=False)
subset["med_house_value"] = subset["med_house_value"].apply(lambda x: x.mid)
```

```python
pp = sns.pairplot(data=subset, hue="med_house_value", palette="viridis")
```

# Feature and Target Variable Selection

```python
x = df.drop("med_house_value", axis=1)
y = df['med_house_value']
```

# Training and Testing Data Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=2)
```

# Linear Regressor Model

```python
from sklearn import linear_model
```

```python
# Linear Regression
linear_reg = linear_model.LinearRegression()
linear_reg.fit(train_x, train_y)
print(linear_reg.score(train_x, train_y))
print(linear_reg.score(test_x, test_y))
```

```python
# Ridge Regression with default parameter
ridge_reg = linear_model.Ridge(alpha=0.1, max_iter = 1000, tol = 0.1)
ridge_reg.fit(train_x, train_y)
```

```python
ridge_reg.score(test_x, test_y)
```

```python
ridge_reg.score(train_x, train_y)
```

```python
# Ridge Regression with alpha = 0.5
ridge_reg = linear_model.Ridge(alpha=0.5, max_iter = 1000, tol = 0.1)
ridge_reg.fit(train_x, train_y)
print("Test Score: ", ridge_reg.score(test_x, test_y))
print("Train Score: ", ridge_reg.score(train_x, train_y))
```

```python
# Ridge Regression with alpha = 0.7
ridge_reg = linear_model.Ridge(alpha=0.7, max_iter = 1000, tol = 0.1)
ridge_reg.fit(train_x, train_y)
print("Test Score: ", ridge_reg.score(test_x, test_y))
print("Train Score: ", ridge_reg.score(train_x, train_y))
```

```python
# Ridge Regression with alpha = 0.05
ridge_reg = linear_model.Ridge(alpha=0.05, max_iter = 100, tol = 0.1)
ridge_reg.fit(train_x, train_y)
print("Test Score: ", ridge_reg.score(test_x, test_y))
print("Train Score: ", ridge_reg.score(train_x, train_y))
```

```python
# Ridge Regression with alpha = 0.01
ridge_reg = linear_model.Ridge(alpha=0.01, max_iter = 100, tol = 0.1)
ridge_reg.fit(train_x, train_y)
print("Test Score: ", ridge_reg.score(test_x, test_y))
print("Train Score: ", ridge_reg.score(train_x, train_y))
```

```python
# Ridge Regression with alpha = 0.005
ridge_reg = linear_model.Ridge(alpha=0.005, max_iter = 1000, tol = 0.1)
ridge_reg.fit(train_x, train_y)
print("Test Score: ", ridge_reg.score(test_x, test_y))
print("Train Score: ", ridge_reg.score(train_x, train_y))
```

```python
# Ridge Regression with alpha = 0.0001
ridge_reg = linear_model.Ridge(alpha=0.0001, max_iter = 100, tol = 0.1)
ridge_reg.fit(train_x, train_y)
print("Test Score: ", ridge_reg.score(test_x, test_y))
print("Train Score: ", ridge_reg.score(train_x, train_y))
```

```python
# Lasso Regression with default values
lasso_reg = linear_model.Lasso(alpha=0.1, max_iter = 1000, tol = 0.1)
lasso_reg.fit(train_x, train_y)
print("Train Score: ", lasso_reg.score(train_x, train_y))
print("Test Score: ", lasso_reg.score(test_x, test_y))
```

```python
# ElasticNet Regression with default values
en_reg = linear_model.ElasticNet(alpha = 1.0, l1_ratio=0.5)
en_reg.fit(train_x, train_y)
print("Train Score: ", en_reg.score(train_x, train_y))
print("Test Score: ", en_reg.score(test_x, test_y))
```

# Model Evaluation

```python
from sklearn.metrics import mean_squared_error
```

```python
# MSE for linear regression
linear_predict = linear_reg.predict(test_x)
linear_mse = mean_squared_error(linear_predict, test_y, squared = False)
linear_mse
```

```python
# MSE for ridge regression
ridge_predict = ridge_reg.predict(test_x)
ridge_mse = mean_squared_error(ridge_predict, test_y, squared=False)
ridge_mse
```

```python
# MSE for lasso regression
lasso_predict = lasso_reg.predict(test_x)
lasso_mse = mean_squared_error(lasso_predict, test_y, squared = False)
lasso_mse
```

```python
# MSE for elastic net regression
en_predict = en_reg.predict(test_x)
en_mse = mean_squared_error(en_predict, test_y, squared=False)
en_mse
```
